## Tuples:-  
its a built data type which allows user to create **immutable** sequence of values.  
its same as list. difference is about **()** and its immutable.  
Tuple allows duplicates also.  
**Strings and tuples are immutable. lists are mutable**

```python
tup = (87,64,33,95,76,33)   #tuple allows duplicates also.
tup
```

```python
tup[3]
```

```python
tup[3] = 23  #tuple does not allow element\item assignment.
```

```python
tup = ()
print(tup)
print(type(tup))
```

```python
tup = (1)
print(tup)
print(type(tup))   #its showing that its type is int. its not correct way to write tuple.
```

```python
tup = (1,)    #this is the right way of writing tuple.
print(tup)
print(type(tup))     
```

**Tuple slicing is same as list slicing.**

### Tuple methods:- 

```python
tup = (1,3,7,2,3,3,6,3,9)
tup
```

```python
tup.count(3)   #it will show count of element 3.
```

```python
tup.index(7)   #will show index of item 7.
```

```python
# wap to check if a list contains a palindrome of elements.
string = input('Enter string to check palindrom: ')
palindrom_string = string[::-1]
if string == palindrom_string:
    print(f'Given {string} is palindrome')
else:
    print(f'Given {string} is not palindrome')
```

```python
# wap to count the number of students with the 'A' grade in the following tuple.
lst = ['C','D','D','A','A','A','B','B','A']
count = lst.count('A')
count1 = lst.count('B')
count2 = lst.count('C')
count3 = lst.count('D')


print(f'Count of A is: {count}')
print(f'Count of B is: {count1}')
print(f'Count of C is: {count2}')
print(f'Count of D is: {count3}')
```
