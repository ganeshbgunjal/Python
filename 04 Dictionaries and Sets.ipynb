### Dictionary:-  
this are used to store values in **key:value** pairs.  
they are unordered, mutable, and dont allow duplicates.  
Dict has no index, thats why its unordered.  
we cant enter 2 keys with same name. it will update first key with new value. 

```python
dict1 = {
    'name' : 'Ganesh',
    'percentage' : 'B+',
    'marks' : [60,58,57]
}
```

```python
dict1
```

```python
dict1['name']
```

```python
dict1['percentage']
```

### Nested Dictionaries:-  
dictionary under dictionary  

```python
student = {
    'name' : 'Ganesh',
    'Subjects' : {
        'Phy' : 97,
        'Chem' : 96,
        'Bio' : 98
    }    
}
```

```python
student
```

```python
student['Subjects']
```

```python
student['Subjects']['Chem']
```

```python
student.keys()
```

```python
student.values()
```

```python
student.items()
```

```python
student.get('name')
```

```python
student.get('Subjects')
```

```python
student.update({'Surname':'Gunjal'})
```

```python
student
```

```python
student.update({'City':'Bangalore'})
```

```python
student
```
