### Strings:- 

Its a data type that stores a sequence of characters.

```python
name = 'Ganesh'   #its a string that stores some characters
print(f'Type is: {type(name)}')
print(f'Id is: {id(name)}')
```

#### Escape Sequencing characters:-  
Their function is to give proper formatting to characters. they are \n,\t etc.  
\n: next line.  
\t: tab space.

```python
str1 = 'This is string first. \nthis is string second. \tThis is string third' 
print(str1)
```

### Basics Operations:-- 

#### 1. String Concatenation:

```python
name = input('Enter name: ')
surname = input('Enter surname: ')
fullname = name +' '+surname
print(f'Full name is: {fullname}')
```

#### 2. Length of String:

```python
str1 = 'This is string first. \nthis is string second. \tThis is string third' 

print('Length of the string is: ',len(str1))
```

#### 3. Indexing  
Indexing starts with zero always in programming. in python index assignment is not allowed. we cant change character in string.

```python
name = 'Ganesh'
print(name[0])
print(name[1])
```

```python
name[2] = 'm'    #string assignment is not allowed.
```

#### 4. Slicing:  
taking piece of string from one position to next specified position

```python
str2 = 'this is another string for slicing'
print(str2[2:7])     #here 2 is inclusive and 7 is exclusive.
```

```python
print(str2[:])   #means print from start to end
```

```python
print(str2[2:])    #from index 2 to end
```

```python
print(str2[:5])  #from start to index 5.
```

```python
# Negative Index
fruit = 'Apple'    #negative index starts with -1 from the end. 
fruit[-4:-2]      #-2 will be exclusive.
```

### Conditional Statements:-  
IF-ELIF-ELSE  
if (condition):  
    statement1  
elif (condition2):  
    statement2  
else:  
    statementN

```python
age = int(input('Enter age: '))
if age <=10:
    print('your are child')
elif age >10 and age <=18:
    print('you are teenager')
elif age >18 and age <=30:
    print('you are adult')
else:
    print('you are old age')
```

```python
light = input('Enter light: ')
if light == 'green':
    print('You can go')
elif light == 'yellow':
    print('you can go but see both sides first')
elif light == 'red':
    print('you cant go')
else:
    print('you entered invalid light')
```

```python
marks = int(input('Enter Marks: '))
if marks >= 90:
    print('Grade A')
elif marks >=80 and marks <90:
    print('Grade B')
elif marks >=70 and marks <80:
    print('Grade C')
elif marks >=60 and marks <70:
    print('Grade D')
else:
    print('You are failed')
```

```python
# wap to check if number entered by user is odd or even.
num = int(input('Enter number: '))
if num % 2 == 0:
    print('Number is Even')
else:
    print('Number is Odd')
```

```python
# wap to find greatest of 3 numbers entered by user.
num1 = int(input('Enter num1: '))
num2 = int(input('Enter num2: '))
num3 = int(input('Enter num3: '))

if (num1>num2) and (num1>num3):
    greatest = num1
elif (num2>num1) and (num2>num3):
    greatest = num2
else:
    greatest = num3
print('Greatest number is: ',greatest)
```
