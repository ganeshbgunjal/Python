## Types of Operators:--- 


### 1. Arithmetic Operators:--- 
that are used for addition substraction multiplication purpose.

```python
a=8
b=3
print(f'Addition: {a+b}')
print(f'Substraction: {a-b}')
print(f'Multiplication: {a*b}')
print(f'Division: {a/b}')
print(f'Remainder: {a%b}')   #modulo operator. to find out remainder.
print(f'Power operator: {a**b}')     #this is power operator. a to the power b.
```

```python
# here a and b are operands and (+) this sign is operator.
```

same for substraction, multiplication, division, and remainder(%).

### Relational/Comparison Operators:--

==,!=,<,>,<=,>= are some operators. they are used for camparing two values like less that or greater than or else. <br>
it returns True or False. mind this.

```python
a=40
b=20
print(a==b)
print(a!=b)
print(a<b)
print(a<=b)
print(a>b)
print(a>=b)
```
